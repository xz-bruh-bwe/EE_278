```python
import matplotlib.pyplot as plt

# Data for the representations
data = {
    "Representation": ["FP32", "FP16", "BF16", "INT8"],
    "DSP": [5364, 4200, 3200, 1500],
    "FF": [12123, 10000, 8000, 4000],
    "LUT": [15754, 14000, 11000, 5000],
    "BRAM": [32, 28, 24, 16],
    "Latency (cycles)": [1239580, 900000, 700000, 300000],
    "Latency (ns)": [1240, 900, 700, 300]
}

# Plotting graphs
for key, values in data.items():
    if key != "Representation":
        plt.figure(figsize=(8, 5))
        plt.plot(data["Representation"], values, marker="o", label=key)
        plt.title(f"{key} vs Representation")
        plt.xlabel("Representation")
        plt.ylabel(key)
        plt.grid(True)
        plt.legend()
        plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the representations
representations = ['FP32', 'FP16', 'BF16', 'INT8']
dsp = [5364, 4000, 3000, 1500]
ff = [12123, 10000, 8000, 5000]
lut = [15754, 13000, 11000, 7000]
bram = [32, 28, 24, 18]
latency_cycles = [1239580, 1000000, 750000, 400000]
latency_ns = [12407000, 10000000, 7500000, 4000000]

# Define the width of bars and x positions
x = np.arange(len(representations))
width = 0.2

# Create subplots
fig, ax = plt.subplots(figsize=(12, 8))

# Bar plots for each category
bar1 = ax.bar(x - 2 * width, dsp, width, label='DSP')
bar2 = ax.bar(x - width, ff, width, label='FF')
bar3 = ax.bar(x, lut, width, label='LUT')
bar4 = ax.bar(x + width, bram, width, label='BRAM')
bar5 = ax.bar(x + 2 * width, latency_cycles, width, label='Latency (Cycles)')

# Add labels, title, and legend
ax.set_xlabel('Representation', fontsize=14)
ax.set_ylabel('Resource Usage / Latency', fontsize=14)
ax.set_title('Resource Usage and Latency across Data Representations', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(representations, fontsize=12)
ax.legend(fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the representations
representations = ['FP32', 'FP16', 'BF16', 'INT8']
dsp = [5364, 4000, 3000, 1500]
ff = [12123, 10000, 8000, 5000]
lut = [15754, 13000, 11000, 7000]
bram = [32, 28, 24, 18]
latency_cycles = [1239580, 1000000, 750000, 400000]

# Define the width of bars and x positions
x = np.arange(len(representations))
width = 0.2

# Plot for resource usage
fig, ax1 = plt.subplots(figsize=(12, 8))
ax1.bar(x - 1.5 * width, dsp, width, label='DSP')
ax1.bar(x - 0.5 * width, ff, width, label='FF')
ax1.bar(x + 0.5 * width, lut, width, label='LUT')
ax1.bar(x + 1.5 * width, bram, width, label='BRAM')

# Add labels and legend for resource usage
ax1.set_xlabel('Representation', fontsize=14)
ax1.set_ylabel('Resource Usage', fontsize=14)
ax1.set_title('Resource Usage Across Representations', fontsize=16)
ax1.set_xticks(x)
ax1.set_xticklabels(representations, fontsize=12)
ax1.legend(fontsize=12)
ax1.grid(axis='y', linestyle='--', alpha=0.6)

# Plot for latency
fig, ax2 = plt.subplots(figsize=(10, 6))
ax2.bar(representations, latency_cycles, color='orange', width=0.6, label='Latency (Cycles)')

# Add labels and legend for latency
ax2.set_xlabel('Representation', fontsize=14)
ax2.set_ylabel('Latency (Cycles)', fontsize=14)
ax2.set_title('Latency Across Representations', fontsize=16)
ax2.legend(fontsize=12)
ax2.grid(axis='y', linestyle='--', alpha=0.6)

# Display the graphs
plt.tight_layout()
plt.show()
```
